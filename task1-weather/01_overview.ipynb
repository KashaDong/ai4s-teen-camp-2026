## 1. 题目概述

天气瓶是一种可爱的装饰品，其源于英国气象学家罗伯特·菲茨罗伊的发明，他于1831年担任达尔文的“小猎犬号”船长时，将蒸馏水、乙醇、樟脑、硝酸钾和氯化铵混合制成溶液密封在玻璃瓶中。这种装置通过溶液结晶形态变化预测天气，例如清澈代表晴朗，浑浊预示多云，大片结晶则预示降温。现代研究发现，天气瓶内的结晶形态其实与实际天气关联性很弱，这些结晶主要是樟脑（硝酸钾和氯化铵作为晶核，促进结晶），无法反映湿度或者气压变化，但与温度变化有关联。虽然天气瓶无法准确预报天气，但因其结晶呈羽毛状或雪花状，十分美观，所以逐渐演变成一种颇受欢迎的装饰品。

科学家在未知星球发现了一种物质$X$，具有与樟脑相似的性质，也是一种易溶于乙醇但难溶于水的固体，通过调节乙醇的质量分数（即乙醇的质量占乙醇加水混合溶剂总质量的百分数），可以改变物质$X$在混合溶剂中的溶解度（每100g溶剂中溶解的溶质的质量，单位：g）。且物质$X$在溶剂中的溶解度，会随着温度升高而显著升高。还有研究表明，当物质$X$的溶解度低于10g（临界值）时，会因结晶速率过快形成帅气的羽毛状结晶，反之则是美丽的雪花状结晶。

```python
def process_and_fill_data(data_path):
    # 用来存放所有处理好的数据
    all_data = {}
    
    # 逐个读取文件夹里的文件
    for filename in os.listdir(data_path):
        # 拼接完整的文件路径
        file_path = os.path.join(data_path, filename)
        
        # 从文件名中提取数字（比如从"5.csv"中提取5）
        file_number = int(filename.split('.')[0])
        
        # 读取文件内容
        # 把表格转置（行变列，列变行）
        df = pd.read_csv(file_path, header=None, na_values=[' ']).T
        # 给两列数据起名字
        df.columns = ["Temperature (K)", "Solubility (g)"]
        # 确保数据是数字格式
        df["Temperature (K)"] = pd.to_numeric(df["Temperature (K)"])
        df["Solubility (g)"] = pd.to_numeric(df["Solubility (g)"])
        
        # 把温度作为索引，溶解度作为值
        solubility_series = df.set_index("Temperature (K)")["Solubility (g)"]
        filled_series = solubility_series.copy()
        
        # 检查并填补缺失的数据点
        for i in range(len(solubility_series)):
            # 如果发现空值
            if pd.isna(solubility_series.iloc[i]):
                # 找到前面最近的一个有效值
                prev_val = solubility_series.iloc[:i].dropna().iloc[-1]
                # 找到后面最近的一个有效值
                next_val = solubility_series.iloc[i+1:].dropna().iloc[0]
                # 用前后两个值的平均数来填补空值
                filled_series.iloc[i] = (prev_val + next_val) / 2
        
        # 把处理好的数据保存
        all_data[file_number] = filled_series.to_dict()
        
    return all_data

# 开始处理数据
print("正在处理数据...")
filled_data = process_and_fill_data(data_path="data_training")

print("--- 所有文件处理完成 ---")
print("处理后的数据：")
print(filled_data)
```

### 子任务 3：识别羽毛状结晶的条件

物质 X 在溶剂中的溶解度决定了结晶形态：
当溶解度 低于 10g 时，结晶速度过快，会形成羽毛状晶体。
当溶解度 大于或等于 10g 时，会形成雪花状晶体。
这一任务中，我们要找出所有能够形成羽毛状晶体的条件，即 温度和乙醇质量分数的组合。首先将表格“展平”成一个长列表，然后进行筛选和排序，从而快速定位符合条件的配方。

#### 3.1 数据“展平”

我们前面整理好的矩阵，行是温度，列是乙醇质量分数。我们要从中筛选出所有小于 10g 的数值。

**代码说明**：二维表格（DataFrame）不方便直接筛选。我们可以使用 .stack() 方法将表格从“宽”格式（行和列）转换为“长”格式（所有数据都在一列中），使其更易于进行条件筛选。

```python
# 使用 .stack() 方法将表格展平，然后用 .reset_index() 将索引转为列。
stacked = solubility_matrix.stack().reset_index()

# 重新命名列，使其更具可读性。
stacked.columns = ["Mass fraction (Ethanol), %", "Temperature (K)", "Solubility (g)"]

print("--- 展平后的数据列表 ---")
print(stacked)
```

#### 3.2 将筛选出的数据整理成表格，并按照溶解度从小到大排序，导出csv文件。

表格的列顺序为：溶解度、温度、乙醇质量分数。

**代码说明**：下面的代码从展平后的数据列表中，筛选出所有溶解度小于 10g 的记录，并按照溶解度值从小到大进行排序。

```python
# 定义溶解度临界值。
critical_value = 10 

# 使用条件筛选，只保留 'Solubility (g)' 小于 10 的行。
feather_recipes_df = stacked[stacked["Solubility (g)"] < critical_value]

# 重新排列列的顺序，让溶解度排在第一列。
feather_recipes_df = feather_recipes_df[['Solubility (g)', 'Temperature (K)', "Mass fraction (Ethanol), %"]]

# 按照溶解度值进行升序排序。
feather_recipes_df = feather_recipes_df.sort_values(by="Solubility (g)", ascending=True)

print("--- 符合羽毛状晶体条件的配方列表 ---")
print(feather_recipes_df)
```

### 子任务 4：可视化——制作溶解度“热力图”

我们将使用热力图（Heatmap）来直观地展示产生羽毛状晶体条件下的溶解度数据。热力图通过颜色的深浅来表示数值大小，颜色越深代表溶解度越高，反之则越低，这能帮助我们快速识别数据趋势。

**代码说明：** 为了规避字体问题，图表中的所有标签都将使用英文。这段代码将使用 seaborn 库绘制热力图，并对图表标题、坐标轴标签、数值格式以及字体大小等细节进行设置，以确保图表清晰。

```python
# 设置画布大小，使图表清晰可见。
# 第一个数字是宽度，第二个数字是高度，单位是英寸。可以尝试： (8,6)小一点, (12,10)大一点, (15,12)更大
plt.figure(figsize=(10, 8))

# 使用 seaborn 库的 .heatmap() 函数绘制热力图。
# 小贴士：每次只修改一个参数，看看图表有什么变化！这样可以更好地理解每个参数的作用。
sns.heatmap(
    solubility_matrix,        # 输入处理好的溶解度表格数据（必须填）
    
    annot=True,               # 在图表上显示具体的数值。
    # 可以换成：True(显示数字), False(不显示数字)
    
    fmt=".4f",                # 设置数值格式，保留4位小数。
    # 可以换成：".1f"(保留1位), ".2f"(保留2位), ".0f"(不要小数)
    
    cmap="viridis",           # 选择一种颜色方案。
    # 可以换成："plasma"(紫红色), "coolwarm"(蓝红对比), "YlOrRd"(黄橙红),"Blues"(蓝色系), "Greens"(绿色系), "RdYlBu"(红黄蓝)
    
    linewidths=.5,            # 设置单元格之间的线宽。
    # 可以换成：0(无线), 1(粗线), 0.2(细线)
    
    linecolor='lightgray',    # 设置线的颜色。
    # 可以换成：'white'(白色), 'black'(黑色), 'red'(红色)
    # 或者颜色代码：'#CCCCCC'(浅灰), '#FF0000'(纯红)
    
    cbar_kws={'label': 'Solubility (g)'}, # 为颜色条添加标签。
    # 可以换成：{'label': '溶解度 (克)', 'shrink': 0.8} (缩短颜色条)
    # 或者：{'label': 'Solubility', 'orientation': 'horizontal'} (横放颜色条)
    
    annot_kws={'fontsize': 6} # 缩小单元格内数值的字体大小，防止溢出。
    # 可以换成：{'fontsize': 8}(大一点), {'fontsize': 10, 'color': 'red'}(红色数字), {'fontsize': 12, 'fontweight': 'bold'}(粗体大字)
)

# 使用英文设置图表的标题。
plt.title("Solubility Heatmap of Substance X", fontsize=16)
# 使用英文设置x轴标签。
plt.xlabel("Temperature (K)", fontsize=12)
# 使用英文设置y轴标签。
plt.ylabel("Ethanol Mass Fraction (%)", fontsize=12)

# 调整x轴标签的角度，以免重叠。
plt.xticks(rotation=45)
# 调整y轴标签的角度。
plt.yticks(rotation=0)

# 自动调整图表布局，确保所有元素完整显示。
plt.tight_layout()

# 显示最终绘制好的图表。
plt.show()
```
